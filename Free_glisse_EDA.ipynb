```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Data_Free_Glisse.csv",sep=';')
```

```python
df=df.iloc[:,1:]
```

```python
df.info()
```

```python
df['Prix']=df['Prix'].str.replace(',','.')
```

```python
df['Prix']=df['Prix'].astype('float')
```
